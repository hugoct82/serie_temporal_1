### Generador de hs por mes


```python
import datetime
import pandas as pd

def calcular_horas_mes(year, month):
    # Obtener el primer día del mes
    primer_dia = datetime.date(year, month, 1)
    
    # Obtener el último día del mes
    if month == 12:
        ultimo_dia = datetime.date(year + 1, 1, 1) - datetime.timedelta(days=1)
    else:
        ultimo_dia = datetime.date(year, month + 1, 1) - datetime.timedelta(days=1)
    
    # Calcular el número de días en el mes
    dias_en_el_mes = (ultimo_dia - primer_dia).days + 1
    
    # Calcular el número de horas en el mes
    horas_en_el_mes = dias_en_el_mes * 24
    
    return horas_en_el_mes

# Crear un DataFrame para almacenar los resultados
fechas = pd.date_range(start='2013-01-01', end='2024-05-31', freq='MS')
horas_por_mes = []

for fecha in fechas:
    year = fecha.year
    month = fecha.month
    horas_mes = calcular_horas_mes(year, month)
    horas_por_mes.append(horas_mes)

# Crear el DataFrame con los resultados
df_horas = pd.DataFrame({
    'Fecha': fechas,
    'Horas': horas_por_mes
})

df_horas
```

```python
csv_url_lats = "https://docs.google.com/spreadsheets/d/1v7iWzENQlawYkBWTJhcYE3u4uoICqSxHSsAq4t68W-8/export?format=csv&id=1v7iWzENQlawYkBWTJhcYE3u4uoICqSxHSsAq4t68W-8&gid=0"
# Cargar el CSV y especificar los separadores de miles y decimales
df = pd.read_csv(csv_url_lats,  decimal=',')# thousands='.',)

df.dtypes
```

```python
# Sumar la columna 'km' en df
suma_km = df['km'].sum()


# Asegúrate de que el índice de df_horas esté en el mismo formato que las fechas de df
df_horas['Fecha'] = pd.to_datetime(df_horas['Fecha'], format='%Y-%m')
df_horas.set_index('Fecha', inplace=True)

# Realiza la multiplicación por cada fila en df_horas
df_horas['Resultado'] = df_horas['Horas'] * suma_km
```

```python
df_horas['Horas_Totales'] = suma_km
```

```python
df_horas.rename(columns={'Resultado': 'uxt_mensual'}, inplace=True)
df_horas.to_csv('./horas_totales_mes.csv', sep=";" , index=True)
```
